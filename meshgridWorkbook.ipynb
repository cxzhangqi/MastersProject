```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set up x, y and z for positions of a (sound source)
x = np.linspace(0, 20, 21).reshape(-1, 1)
y = np.zeros_like(x)
z = np.zeros_like(x)

# Create an (number_points x 3) array of the 'a' positions
a = np.concatenate((x, y, z), axis=1)
H = np.array([[0,0,0],[20, 0, 0]],dtype=float)
```

```python
# Setting up the points in the x-y plane
x = np.linspace(0,20,21)
y = np.linspace(0,20,21)
X, Y = np.meshgrid(x,y,indexing='xy')
Y = Y.reshape(-1,1)
X = X.reshape(-1,1)
Z = np.zeros_like(X)


a = np.concatenate((X, Y, Z), axis=1)
a
```

```python
def distance_multimic(a, H):
    """

    :param a: 1 x 3 array
    :param H: i x 3 array
    :return: 1 x i array of distance values
    """
    out = np.zeros((a.shape[0], H.shape[0]))
    
    for i in range(H.shape[0]):
        out[:, i] = np.sqrt(np.sum(np.square(H[i, :] - a), axis=1))
    return out
```

```python
dist = distance_multimic(a,H)
dist[:,0].reshape(21,-1)
```

What is happening above is that we are changing the meshgrid into a 1D array so that we can do calculations on it easily. Then when we have calculated what we want through the distance calculations, we SHOULD be able to just reshape back to our meshgrid to obtain Z.
